```python
#metadata upload
intime=[]
for idx1,i in enumerate(metadata['urlid']):
    try:
        st=metadata['st'][idx1]
        st=datetime.strptime(st, "%Y-%m-%dT%H:%M:%S.%fZ")
        et=metadata['et'][idx1]
        et=datetime.strptime(et, "%Y-%m-%dT%H:%M:%S.%fZ")
        path='../tsvt/chat/{}''\n''.txt'.format(i)
        chat=pd.read_csv(path,sep='\t',names=['leage','game','view','length','real_time','duration','chat'])
        s_idx=0
        e_idx=0
        temp=[]
        diff=0


        for idx,j in enumerate(chat['real_time']):
            chat_time=datetime.strptime(j, "%Y-%m-%d %H:%M:%S")-timedelta(hours=9) #original과 시간 맞춤

            if chat_time>st:
                if s_idx==0:
                    diff=(chat_time-st).seconds
                    s_idx=idx
                    temp.append(diff)
                    temp.append(chat['duration'][idx])
                else:
                    pass
            if chat_time>et:
                if e_idx==0:
                    e_idx=idx-1
                    temp.append(chat['duration'][e_idx])
                    break
        intime.append(temp)
        chat.iloc[s_idx:e_idx].to_csv('./result/{}.csv'.format(str(metadata['gameid'][idx1])))
    
    except:
        print(metadata['gameid'][idx1])
        print('error : '+str(idx1))
        intime.append(['','',''])
        pass



```
