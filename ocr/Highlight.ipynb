```python
import pandas as pd
```

```python
data=pd.read_csv('labeling (11).csv')
```

```python
data['102979081290790284 data']
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
data.columns
```

```python
import datetime
label_result={}
for i in list(data.columns):
    if "data" in i:
        #print(i)
        time_list=[]
        sp=0
        for idx,time in enumerate(data[i]):
            if ":" in str(time):
                next=data[i][idx+1]
                if (int(next.split(':')[0])*60+int(next.split(':')[1])*1)-(int(time.split(':')[0])*60+int(time.split(':')[1])*1)<0:
                    print("==={}{} there are some error\n".format(i,next))
                    pass
                if ":" not in str(data[i][idx+2]): #마지막
                    time_list.append([sp,next])
                    break
                
                if sp==0:#new 구간
                    sp=time

                if (int(next.split(':')[0])*60+int(next.split(':')[1])*1)-(int(time.split(':')[0])*60+int(time.split(':')[1])*1)>3: # 다음꺼-현재>3 && 다다음꺼-다음꺼<3
                    if (int(data[i][idx+2].split(':')[0])*60+int(data[i][idx+2].split(':')[1])*1)-(int(next.split(':')[0])*60+int(next.split(':')[1])*1)<3 and (int(data[i][idx+2].split(':')[0])*60+int(data[i][idx+2].split(':')[1])*1)-(int(next.split(':')[0])*60+int(next.split(':')[1])*1)>=0:
                        time_list.append([sp,time])
                        sp=0
                    else:
                        #뭔가 문제가 있는애들 체크
                        print(i)
                        print(time)
                        print(next)
                        print(data[i][idx+2])
                        pass
        label_result[i.split()[0]]=time_list
                
#시간이 역순인애들
# cur_point / cur_point+1 차이가 3 && cur_point+1 / cur_point+2 차이가 3  [하이라이트가 단 1초인경우]
```

```python
for i in label_result.keys():#시간 역순인애들 캐치 못한 경우
    for x,y in label_result[i]:
        if int(y.split(':')[0])*60+int(y.split(':')[1])<int(x.split(':')[0])*60+int(x.split(':')[1]):
            print(x,y)
            print(i)
```
