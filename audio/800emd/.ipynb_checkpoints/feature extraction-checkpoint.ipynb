# features

### 2차원 array로 EMD signal 저장, array 하나 단위당 800sampe, 1초당 800sample
### 즉 1초단위로 EMD

#### energy(frame) = S|sample limits|^2

### 전체 길이는 feature_1 기준으로

```python
from matplotlib import pyplot as plt
import numpy as np
import os
import pickle
```

```python
path = '/home/ubuntu/gangmin/emd_800'
file_list = os.listdir(path)
```

```python
features_1 ={}
audio_energy_1={}
```

```python
#800sampel 1 shot에 80sample 10frame들어가서 측정, 즉 초*10frame energy생성후 상위 15% 추출해서 아 frame이 속하는 shot 모두 선택
for file in file_list:
    data=[]
    with open(path+'/'+file,'rb') as f:
        shots = pickle.load(f)
        energy = []
        for shot in shots:
            index = 0
            for frame_index in range(80,len(shot),80):
                energy.append(np.sum(np.square(np.abs(shot[index:frame_index]))))
                index = frame_index
            energy.append(np.sum(np.square(np.abs(shot[index:len(shot)]))))

    features_1[file[:-7]] = np.zeros(int(len(energy)/10))
    
    audio_energy_1[file[:-7]] = energy
    
    tail = int((len(energy)/100)*15)
    
    
    top = sorted(range(len(energy)), key=lambda i: energy[i])[-tail:]
    
    for i in range(int(len(energy))):
        if i in top:
            features_1[file[:-7]][i//10] = 1
```

```python
features_3 ={}
audio_energy_3={}
```

```python
#shot 내에서 초당 averaged된 energy의 증가율 - 50%로 설정, 대안은 100%
for file in file_list:
    data=[]
    with open(path+'/'+file,'rb') as f:
        shots = pickle.load(f)
        energy = []
        for shot in shots:
            energy.append(np.average(shot))
        audio_energy_3[file[:-7]] = energy  
        features_3[file[:-7]] = np.zeros(len(features_1[file[:-7]]))
        top = sorted(range(len(energy)), key=lambda i : energy[i])[-int(len(energy)/100*15):]
        
        for i in top:
            features_3[file[:-7]][i] = 1
        
            
```

```python
import pickle
with open('/home/ubuntu/gangmin/tsvt/multimodal/data/audio_energy_3.pickle','wb')as f:
    pickle.dump(audio_energy_3,f)
```
