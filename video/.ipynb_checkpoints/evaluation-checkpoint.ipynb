```python
import json

with open('/home/ubuntu/gangmin/tsvt/share_data/label.json') as json_file:
    label = json.load(json_file)
    
```

```python
label
```

```python
labels = {}
```

```python
import numpy as np

for key in label.keys():
    m = label[key][-1][-1].split(':')[0]
    s= label[key][-1][-1].split(':')[1]
    temp = np.zeros(int(m)*60+int(s)+1)
    for interval in label[key]:
        t=[]
        for time in interval:
            m = time.split(':')[0]
            s = time.split(':')[1]
            m = int(m)*60
            s = int(s)
            t.append(m+s)
        for time in range(t[0],t[1]+1):
            temp[time] = 1
    
    labels[key] = temp
```

```python
labels
```

```python
import pickle
with open('/home/ubuntu/gangmin/tsvt/video/lolData/exp_data/highlight_label.pickle',"wb") as f:
    pickle.dump(labels,f)
```

```python
for key in label.keys():
    print(len(labels[key]))
    frames[key] = frames[key][:len(labels[key])]
    print(len(frames[key]))
```

```python
for key in label.keys():
    for end in range(len(frames[key])-27,len(frames[key])):
        frames[key][end] = 1
```

# TP FP TN FN

```python
average = [0,0,0,0,0]
for key in labels.keys():
    TP = len(np.where((frames[key]==1)&(labels[key]==1)==True)[0])
    FP = len(np.where((frames[key]==1)&(labels[key]==0)==True)[0])
    TN = len(np.where((frames[key]==0)&(labels[key]==0)==True)[0])
    FN = len(np.where((frames[key]==0)&(labels[key]==1)==True)[0])

    
    precision = TP/(TP+FP)
    recall = TP/(TP+FN)
    accuracy = (TP+TN) / (TP+FN+FP+TN)
    
    if precision==0 and recall == 0:
        print('!')
    else:
    
        f1 = (2*precision*recall / (precision + recall))
    
        print('precision : {}, recall : {}, f1 : {}, accuracy : {}'.format(precision,recall,f1,accuracy))
        average[0]+= precision
        average[1] += recall
        average[2] += f1
        average[3] += accuracy
        average[4]+=1
```

```python
average
```

```python
print('precision : {}, recall : {}, f1 : {}, accuracy : {}'.format(average[0]/average[4],average[1]/average[4],average[2]/average[4],average[3]/average[4]))
```
