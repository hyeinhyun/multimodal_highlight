# 최초 0데이터 제거

```python
for k,v in whole_data.items():
    print(v['frame'].values[0]['blueTeam']['totalGold'])
```

```python
for k in whole_data.keys():
    if whole_data[k]['frame'].values[0]['blueTeam']['totalGold'] == 0:
        whole_data[k]= whole_data[k].drop([whole_data[k].index[0]])
```

```python
whole_data
```

```python
for game_id in whole_data.keys():
    whole_data[game_id]['Timestamp'] = list(map(lambda x:x['rfc460Timestamp'],whole_data[game_id]['frame'].values))
    
    whole_data[game_id]['blue_gold'] = list(map(lambda x:x['blueTeam']['totalGold'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['blue_towers'] = list(map(lambda x:x['blueTeam']['towers'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['blue_barons'] = list(map(lambda x:x['blueTeam']['barons'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['blue_totalKills'] = list(map(lambda x:x['blueTeam']['totalKills'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['blue_dragons'] = list(map(lambda x:len(x['blueTeam']['dragons']),whole_data[game_id]['frame'].values))

    whole_data[game_id]['red_gold'] = list(map(lambda x:x['redTeam']['totalGold'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['red_towers'] = list(map(lambda x:x['redTeam']['towers'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['red_barons'] = list(map(lambda x:x['redTeam']['barons'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['red_totalKills'] = list(map(lambda x:x['redTeam']['totalKills'],whole_data[game_id]['frame'].values))
    whole_data[game_id]['red_dragons'] = list(map(lambda x:len(x['redTeam']['dragons']),whole_data[game_id]['frame'].values))
    
    whole_data[game_id].drop(['frame'],axis='columns',inplace=True)
    whole_data[game_id] = whole_data[game_id].reset_index(drop=True)
```

```python
import pickle
 
## Save pickle
with open("C:/Users/Gangmin/Desktop/gangmin/capstone/lolDeveloperApi/gamedata.pickle","wb") as fw:
    pickle.dump(whole_data, fw)
```
