```python
filelist = ['/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2018-spring-match-data-OraclesElixir-2018-05-20.xlsx',
'/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2018 summer match data OraclesElixir 2018-09-27.xlsx',
'/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2018-worlds-match-data-OraclesElixir-2018-11-03.xlsx',
'/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2019-spring-match-data-OraclesElixir-2019-05-21.xlsx',
'/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2019-summer-match-data-OraclesElixir-2019-09-16.xlsx',
'/home/ubuntu/gangmin/tsvt/video/lolData/leagueoflegends2018~2020/2019-worlds-match-data-OraclesElixir-2019-10-20.xlsx']
game_summary = pd.DataFrame(columns= ['split','goldblue','goldred','bTowers','rTowers','bKills','rKills','bResult','rResult'])
index = 0
for filename in filelist:
    exdata = pd.read_excel(filename)
    
    
    for i in range(len(exdata)):
        if int(exdata.loc[[i]]['playerid'])==100:
            try:
                game_summary.loc[index] = [str(exdata.loc[[i]]['split']),int(exdata.loc[[i]]['totalgold']),int(exdata.loc[[i+1]]['totalgold']),int(exdata.loc[[i]]['teamtowerkills']),int(exdata.loc[[i+1]]['teamtowerkills']),int(exdata.loc[[i]]['teamkills']),int(exdata.loc[[i+1]]['teamkills']),int(exdata.loc[[i]]['result']),int(exdata.loc[[i+1]]['result'])]
                index +=1
            except:
                pass
        
```
