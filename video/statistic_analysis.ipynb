```python
import json

with open('/home/ubuntu/gangmin/tsvt/share_data/label.json') as json_file:
    label = json.load(json_file)
    
```

```python
label['102844341908026005'][-1] = ['48:17', '48:36']
```

```python
labels = {}
```

```python
import numpy as np

for key in label.keys():
    m = label[key][-1][-1].split(':')[0]
    s= label[key][-1][-1].split(':')[1]
    temp = np.zeros(int(m)*60+int(s)+1)
    for interval in label[key]:
        t=[]
        for time in interval:
            m = time.split(':')[0]
            s = time.split(':')[1]
            m = int(m)*60
            s = int(s)
            t.append(m+s)
        for time in range(t[0],t[1]+1):
            temp[time] = 1
    
    labels[key] = temp
```
